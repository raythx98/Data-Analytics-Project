## Obtaining a data of all relevant modules
#### We're going to need a function which cleans the data input, because not everyone follows instructions
<hr>

```python
# Changes all input to upper case, removes leading and trialing spaces, 
# splits all module codes into a list of module codes, then removes all white spaces
def clean(data):
    s0 = data.upper()
    s1 = s0.strip()
    s2 = s1.split(',')
    s3 = []
    for i in s2:
        s3.append(i.replace(' ', ''))
    return s3
# Removes all duplicates, removes all empty inputs 
def remove_anomaly(xs):
    ys = []
    for i in xs:
        if i not in ys:
            ys.append(i)
    return [j for j in ys if j != '']
print("For the inputs of modules, separate them with a comma ','")
print("if your module has different codes for different cohort")
print("it is best if you enter all (e.g cs1231s and cs1231,")
print("ger1000 and ger1000h).")
input ("reply yes if understood\n")
print('')
```

```python
print ("List all modules in your degree requirement")
print ("+ UE/GE modules you want to take")
total = remove_anomaly(clean(input('')))
print('')
while True:
    print(total)
    print("^ These are what we have so far, ")
    print("rerun the cell if there are wrong entries, ")
    print("key in additional modules if you need to, ")
    ans = input('reply "done" to continue\n')
    print('')
    if ans.lower() == 'done':
        break
    else:
        total += clean(ans)
        total = remove_anomaly(total)
        if 'NONE' in total:
            total.remove('NONE')
print('Total Modules (Final): ', total)
```

```python
print("What Modules have you taken so far?")
print('reply "none" otherwise')
taken = remove_anomaly(clean(input('')))
print('')
while True:
    print(taken)
    print("^ These are the modules you have taken so far, ")
    print("rerun the cell if there are wrong entries, ")
    print("key in additional modules if you need to,")
    ans = input('reply "done" to continue\n')
    print('')
    if ans.lower() == 'done':
        break
    else:
        taken += clean(ans)
        taken = remove_anomaly(taken)
        if 'NONE' in taken:
            taken.remove('NONE')
print('Modules Taken (Final): ', taken)
```

```python
print("What Modules do you intend to take during SEP? ")
print('reply "none" otherwise')
intended = remove_anomaly(clean(input('')))
print('')
while True:
    print(intended)
    print("^ These are what we have so far, ")
    print("rerun the cell if there are wrong entries, ")
    print("key in additional modules if you need to,")
    ans = input('reply "done" to continue\n')
    print('')
    if ans.lower() == 'done':
        break
    else:
        intended += clean(ans)
        intended = remove_anomaly(intended)
        if 'NONE' in intended:
            intended.remove('NONE')
print('Final Modules Intending to Take: ', intended)
```

```python
print('What modules are a "must take" for you during SEP?')
print("/ do you intend to take for the whole of NUS?")
print('reply "none" otherwise')
required = remove_anomaly(clean(input('')))
print('')
while True:
    print(required)
    print("^ These are what we have so far, ")
    print("rerun the cell if there are wrong entries, ")
    print("key in additional modules if you need to,")
    ans = input('reply "done" to continue\n')
    print('')
    if ans.lower() == 'done':
        break
    else:
        required += clean(ans)
        required = remove_anomaly(required)
        if 'NONE' in required:
            required.remove('NONE')
print('Final Modules Intending to Take: ', required)
```

#### We need to handle these data
- [taken], [intended], [required] needs to be a subset of [total]
- [taken] is mutually exclusive with [intended] and [required], [taken] takes precedence
- [required] is a subset of [intended]

```python
final_taken = taken
final_required = []
final_intended = []
final_total = []
final_eval = []
data_inputs = [total, taken, intended, required]

final_inputs = [final_total, final_taken
                , final_intended, final_required, final_eval]

# Fill final_total
for each in data_inputs:
    for i in each:
        if i not in final_inputs[0]:
            final_inputs[0].append(i)

# mutually exclusive for taken and intended            
for i in data_inputs[2]:
    if i not in final_inputs[1]:
        final_inputs[2].append(i)

# mutually exclusive for taken and required
for i in data_inputs[3]:
    if i not in final_inputs[1]:
        final_inputs[3].append(i)
        
for i in final_inputs[3]:
    if i not in final_inputs[2]:
        final_inputs[2].append(i)
        
for i in final_inputs[0]:
    if i not in final_inputs[1]:
        final_inputs[4].append(i)
        
print(final_inputs)
```

## Finally, we're ready for the magic
- Now we apply a groupby function on
    - [intended]
    - [total - taken]
    
### Recall desired headings
- Mappings available (based on intended modules) [intended]
- Number of intended mappings
- Mappings available (all applicable) [total - taken] 
- Number of applicable mappings

```python
def calc_maps(df):
    return len(df)
def pair(df):
    return (df[0] + " maps to " + df[1])
```

<br><br><br><br><br><br>

### [Intended]

```python
df_required_copy = df_clean_required.copy()
df_filter_intended = df_required_copy[df_required_copy[
    'NUS Module Code'].isin(final_inputs[2])]
df_filter_intended.sort_values('Host University', inplace = True)
df_filter_intended.reset_index(drop = True, inplace = True)
if not df_filter_intended.empty:
    df_filter_intended[
        "Mappings available (based on intended modules)"
    ] = df_filter_intended[
        ['HU Module Code', 'NUS Module Code']].apply(pair, axis = 1)
    df_filter_intended.drop(
        ['HU Module Code', 'NUS Module Code'], 
        axis = 1, 
        inplace = True)
    df_groupby_intended = df_filter_intended.groupby(
        'Host University'
    )['Mappings available (based on intended modules)'].apply(
        lambda group_series: group_series.tolist()).reset_index()
    df_groupby_intended[
        'Number of intended mappings'
    ] = df_groupby_intended[
        'Mappings available (based on intended modules)'
    ].apply(calc_maps)
    df_groupby_intended.sort_values(
        'Number of intended mappings', 
        inplace = True, 
        ascending = False)
    df_groupby_intended.reset_index(drop = True, inplace = True)
    display(df_groupby_intended.head())
else:
    df_groupby_intended = pd.DataFrame(
        columns=['Host University', 
                 'Mappings available (based on intended modules)', 
                 'Number of intended mappings'])
    display(df_groupby_intended.head())
```

<br><br><br><br><br><br>

### [Total - Taken] 

```python
df_required_copy = df_clean_required.copy()
df_filter_all = df_required_copy[
    df_required_copy['NUS Module Code'].isin(final_inputs[4])]
df_filter_all.sort_values('Host University', inplace = True)
df_filter_all.reset_index(drop = True, inplace = True)
if not df_filter_all.empty:
    df_filter_all[
        "Mappings available (all applicable)"
    ] = df_filter_all[['HU Module Code', 'NUS Module Code']].apply(
        pair, axis = 1)
    df_filter_all.drop(['HU Module Code', 'NUS Module Code'], 
                       axis = 1, inplace = True)
    df_groupby_all = df_filter_all.groupby('Host University')[
        'Mappings available (all applicable)'
    ].apply(
        lambda group_series: group_series.tolist()).reset_index()
    df_groupby_all['Number of applicable mappings'] = df_groupby_all[
        'Mappings available (all applicable)'].apply(calc_maps)
    df_groupby_all.sort_values('Number of applicable mappings', 
                               inplace = True, ascending = False)
    df_groupby_all.reset_index(drop = True, inplace = True)
    display(df_groupby_all.head())
else:
    df_groupby_all = pd.DataFrame(
        columns=['Host University', 
                 'Mappings available (all applicable)', 
                 'Number of applicable mappings'])
    display(df_groupby_all.head())
```

<br><br><br><br><br><br><br><br><br><br><br><br>

## Putting them all together

```python
df_final = pd.merge(
    df_groupby_all, 
    df_groupby_intended, 
    left_on = 'Host University', 
    right_on = 'Host University', 
    how = 'left')
df_final.fillna(int(0), inplace = True)
df_final.sort_values(['Number of intended mappings', 
                      'Number of applicable mappings'
                     ], inplace = True, ascending = [False, False])
df_final.reset_index(drop = True, inplace = True)
order = ['Host University',
 'Mappings available (all applicable)',
 'Number of applicable mappings',
 'Mappings available (based on intended modules)',
 'Number of intended mappings']
df_final = df_final[order]
df_final['Number of intended mappings'] = df_final[
    'Number of intended mappings'].apply(np.int)
df_final.head()
```
